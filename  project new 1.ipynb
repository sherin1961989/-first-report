#How many are there in each Team and the percentage splitting with respect to the total employees:

    

```python
team_counts = df['Team'].value_counts()
team_percentages = (team_counts / len(df)) * 100

print("Number of employees in each Team:")
print(team_counts)

print("\nPercentage splitting with respect to the total employees:")
print(team_percentages)
```

Find from which age group most of the employees belong to:

```python
# Define age groups
bins = [20, 30, 40, 50, 60, 70]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69']

# Categorize employees into age groups
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Find the age group with the most employees
most_common_age_group = df['Age_Group'].value_counts().idxmax()

print("Age group with the most employees:", most_common_age_group)
```

4.Find out under which team and position, spending in terms of salary is high.


```python
# Calculate total spending on salary for each team and position
spending_by_team_position = df.groupby(['Team', 'Position'])['Salary'].sum()

# Find the team and position with the highest spending on salary
highest_spending_team_position = spending_by_team_position.idxmax()
highest_spending_salary = spending_by_team_position.max()
print("spending_by_team_position:",spending_by_team_position)
print("Team and position with the highest spending on salary:", highest_spending_team_position)
print("Highest spending on salary:", highest_spending_salary)
```
